```python
import pandas as pd
import geopandas as gpd
from matplotlib import pyplot as plt
from shapely.geometry import Point
%matplotlib inline
```

```python
taxiJan = pd.read_csv('yellow_tripdata_2016-01.csv')
taxiJan.head()
```

```python
taxiShape = gpd.read_file('taxi_zones/taxi_zones.shp')
taxiShape.head()
```

## Taxi shapes for areas aroung second avenue subway

```python
Id = [140, 141, 262, 263, 75, 236, 237]
taxiShapeNew = taxiShape[(taxiShape.LocationID.isin(Id))]
taxiShapeNew.head()
```

```python
geometry = [Point(xy) for xy in zip(taxiJan.pickup_longitude, taxiJan.pickup_latitude)]
crs = {'init': 'epsg:4326'}
taxiJan = gpd.GeoDataFrame(taxiJan, crs=crs, geometry=geometry)
```

```python
taxiJan.to_crs(epsg=2263, inplace=True)
taxiShapeNew.to_crs(epsg=2263, inplace=True)
```

```python
taxiJan.crs
```

```python
taxiJan1 = gpd.sjoin(taxiJan, taxiShapeNew)
```
